```python
import math

calculate = input("""What do you want to calculate?
type "n" for number of monthly payments,
type "a" for annuity monthly payment amount,
type "p" for loan principal:
""")

if (calculate == "n"):
    loan_principal = int(input("Enter the loan principal:\n"))
    annual_payment = float(input("Enter the monthly payment:\n"))
    loan_interest = float(input("Enter the loan interest:\n"))
    
    i = loan_interest / 100 / 12
    
    month = math.ceil(math.log(annual_payment / (annual_payment - i * loan_principal), 1 + i))
    
    years = month // 12
    month = month % 12
    
    if (years < 1):
        print("It will take " + str(month) + (" month" if month == 1 else " months") + " to repay the loan")
    else:
        print("It will take " + str(years) + (" year and" if years == 1 else " years and ") + str(month) + (" month" if month == 1 else " months") + " to repay this loan!")

elif (calculate == "p"):
    annual_payment = float(input("Enter the annuity payment:\n"))
    periods = int(input("Enter the number of periods:\n"))
    loan_interest = float(input("Enter the loan interest:\n"))
    
    i = loan_interest / 100 / 12
    z = math.pow(1 + i, periods)
    
    loan_principal = math.ceil(annual_payment / ((i * z) / (z - 1)))
    
    print("Your loan principal = " + str(loan_principal) + "!")        

elif (calculate == "a"):
    loan_principal = int(input("Enter the loan principal:\n"))
    periods = int(input("Enter the number of periods:\n"))
    loan_interest = float(input("Enter the loan interest:\n"))
    
    i = loan_interest / 100 / 12
    z = math.pow(1 + i, periods)
    annual_payment = math.ceil(loan_principal * i * z / (z - 1))
    
    print("Your monthly payment = " + str(annual_payment) + "!")
```

```python
import argparse
import math

parser = argparse.ArgumentParser()

parser.add_argument("--type")
parser.add_argument("--principal", type=float)
parser.add_argument("--periods", type=int)
parser.add_argument("--interest", type=float)
parser.add_argument("--payment", type=float)

args = parser.parse_args()

if args.type == "annuity":
    if args.principal and args.periods and args.interest:
        i = args.interest / 100 / 12
        z = math.pow(1 + i, args.periods)
        annuity = math.ceil(args.principal * i * z / (z - 1))

        overpayment = args.principal - annuity * args.periods

        print("Your annuity payment = " + str(int(annuity)) + "!")
        print("Overpayment = " + str(int(abs(overpayment))))

    elif args.payment and args.principal and args.interest:
        i = args.interest / 100 / 12

        months = math.ceil(
                  math.log(
                    args.payment / (args.payment - i * args.principal), 1 + i
                ))

        years = int(months // 12)
        month = int(months % 12)

        if (years < 1):
            print("It will take " + str(month) + (" month" if month == 1 else " months") + " to repay the loan")
        elif (month == 0):
            print("It will take " + str(years) + (" year" if years == 1 else " years") + " to repay the loan")
        else:
            print("It will take " + str(years) + (" year and" if years == 1 else " years and ") + str(month) + (" month" if month == 1 else " months") + " to repay this loan!")

        overpayment = args.principal - args.payment * months 

        print("Overpayment = " + str(int(abs(overpayment))))

    elif args.payment and args.periods and args.interest:
        i = args.interest / 100 / 12
        z = math.pow(1 + i, args.periods)

        principal = math.floor(args.payment / ((i * z) / (z - 1)))

        overpayment = principal - args.payment * args.periods

        print("Your loan principal = " + str(int(principal)) + "!")
        print("Overpayment = " + str(int(abs(overpayment))))

    else:
        print("Incorrect parameters")

elif args.type == "diff":
    if args.principal and args.periods and args.interest:
        i = args.interest / 100 / 12
        z = math.pow(1 + i, args.periods)

        overpayment = args.principal

        for m in range(1, args.periods + 1):
            D = math.ceil(args.principal / args.periods + i * (args.principal - (args.principal * (m - 1)) / args.periods))
            print("Month " + str(m) + ": payment is " + str(int(D)))

            overpayment -= D

        print("Overpayment = " + str(int(abs(overpayment))))

    else:
        print("Incorrect parameters")

else:
    print("Incorrect paramater")
```
